### Baseline linear regression

```python
def CV(model, X, y, cv):
    fitted = model.fit(X,y)
    CV_score = np.mean(cross_val_score(estimator = fitted, X = X, y = y, cv = cv))
    print('Mean CV score for 5 folds = {}'.format(CV_score))
```

```python
def evaluate_model_score(pipeline, train, test, train_y, test_y):
    pipeline.fit(train, train_y)
    pred_test = pipeline.predict(test)
    pred_train = pipeline.predict(train)

    return pd.DataFrame({
        'train_RMSE': [np.sqrt(mean_squared_error(train_y, pred_train))],
        'train_R2': [r2_score(train_y, pred_train)],
        'train_MAE': [mean_absolute_error(train_y, pred_train)],
        'test_RMSE': [np.sqrt(mean_squared_error(test_y, pred_test))],
        'test_R2': [r2_score(test_y, pred_test)],
        'test_MAE': [mean_absolute_error(test_y, pred_test)]
    })
```

### Building model and pipeline 

```python
def build_model():
    
    model = keras.Sequential([
        layers.Dense(64, activation='relu', input_shape=[mapper.fit_transform(train_X).shape[1]]),
        layers.Dense(64, activation='relu'),
        layers.Dense(1)
    ])

    optimizer = tf.keras.optimizers.SGD(learning_rate=0.001)

    model.compile(loss='mse',
                optimizer=optimizer,
                metrics=['mae', 'mse'])
    
    return model
```

```python
import torch 
```

```python
class SineNet(torch.nn.Module):
    def __init__(self, n_features, n_hidden_neurons):
        super(SineNet, self).__init__()
        self.fc1 = torch.nn.Linear(n_features, n_hidden_neurons) # full connected 
        self.act1 = torch.nn.ReLU()
        self.fc2 = torch.nn.Linear(n_hidden_neurons, n_hidden_neurons) # full connected 
        self.act2 = torch.nn.ReLU()
        self.fc3 = torch.nn.Linear(n_hidden_neurons, 1) 

    def forward(self, x):
        x = self.fc1(x)
        x = self.act1(x)
        x = self.fc2(x)
        x = self.act2(x)
        x = self.fc3(x)
        return x

    
sine_net = SineNet(4, 64)
```

```python
optimizer = torch.optim.Adam(sine_net.parameters(), lr=0.001) 
```

```python
def loss(pred, target):
    squares = (pred - target) ** 2
    return squares.mean()

def on_epoch_end(epoch):
    print('.', end='')
    if epoch % 100 == 0: 
        print('')    
```

```python
train_X_NN, train_y_NN = mapper.fit_transform(train_X), train_y
train_X_NN, train_y_NN = torch.from_numpy(train_X_NN).float(), torch.tensor(train_y_NN)

test_X_NN, test_y_NN = mapper.transform(test_X), test_y
test_X_NN, test_y_NN = torch.from_numpy(test_X_NN).float(), torch.tensor(test_y_NN)
```

```python
EPOCHS = 300

for epoch_index in range(EPOCHS):
    optimizer.zero_grad() 
    y_pred = sine_net.forward(train_X_NN) 
    loss_val = loss(y_pred, train_y_NN)
    loss_val.backward()
    optimizer.step()
    on_epoch_end(epoch_index)
    
```

```python
def get_result(model, data):
    data = model.forward(data) 
    data = data.reshape(data.shape[0])
    data = data.detach().numpy()
    return data

def eval_score_torch(model, train_X, train_y, test_X, test_y): 
    pred_test = get_result(model, test_X)
    pred_train = get_result(model, train_X)

    return (pd.DataFrame({
        'train_RMSE': [np.sqrt(mean_squared_error(train_y, pred_train))],
        'train_R2': [r2_score(train_y, pred_train)],
        'train_MAE': [mean_absolute_error(train_y, pred_train)],
        'test_RMSE': [np.sqrt(mean_squared_error(test_y, pred_test))],
        'test_R2': [r2_score(test_y, pred_test)],
        'test_MAE': [mean_absolute_error(test_y, pred_test)]
    }))
```

```python
eval_score_torch(sine_net, train_X_NN, train_y, test_X_NN, test_y)
```

So the score is lower than on the other methods. I can explain it by the fact that when we transform np.array to tenzor, we will get rounded values, so accuracy will decrease...

```python
class MNISTNet(torch.nn.Module):
    def __init__(self, n_input_neurons, n_output_neurons, n_hidden_neurons):
        super(MNISTNet, self).__init__()
        self.fc1 = torch.nn.Linear(n_input_neurons, n_hidden_neurons)
        self.ac1 = torch.nn.ReLU()
        self.fc2 = torch.nn.Linear(n_hidden_neurons, n_hidden_neurons)
        self.ac2 = torch.nn.ReLU()
        self.fc3 = torch.nn.Linear(n_hidden_neurons, n_output_neurons) 
    
    def forward(self, x):
        x = self.fc1(x)
        x = self.ac1(x)
        x = self.fc2(x)
        x = self.ac2(x)
        x = self.fc3(x)
        return x
    
mnist_net = MNISTNet(X_train_NN.shape[1], 4, 50)
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu') 
mnist_net = mnist_net.to(device)
mnist_net.parameters
```

```python
def train_model(model, loss_func, optimizer, num_epochs=10):
    # stats
    epoch_loss = []
    epoch_acc  = []
    
    model.to(device)

    for epochs in range(num_epochs):
        # for stats
        running_loss = 0.0
        running_acc  = 0.0

        for inputs, labels in tqdm(train_dataloader):
            inputs, labels = inputs.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = loss_func(outputs, labels)
            loss.backward() # operation back propagation
            optimizer.step()

            # stats
            running_loss += loss.item()
            running_acc  += (outputs.argmax(dim=1) == labels.data).float().mean()

        epoch_loss.append(running_loss / len(train_dataloader))
        epoch_acc.append(running_acc / len(train_dataloader))

        print(f"Epoch {epochs} - Training loss: {epoch_loss[epochs]} - Accuracy: {epoch_acc[epochs]}")
        
    return epoch_loss, epoch_acc
```

## Building our own model

```python
class OlegNet5(torch.nn.Module):
    def __init__(self):
        super(OlegNet5, self).__init__() 
        
        self.conv1 = torch.nn.Conv2d(
            in_channels=3, out_channels=6, kernel_size=5, padding=2) # 224 ->(pad) 228 -> 224
        self.act1  = torch.nn.ReLU()
        self.pool1 = torch.nn.MaxPool2d(kernel_size=2, stride=2) # 224 -> 112
       
        self.conv2 = torch.nn.Conv2d(
            in_channels=6, out_channels=16, kernel_size=5, padding=0) # 112 -> 108
        self.act2  = torch.nn.ReLU()
        self.pool2 = torch.nn.MaxPool2d(kernel_size=2, stride=2) # 108 -> 54
        
        self.conv3 = torch.nn.Conv2d(
            in_channels=16, out_channels=32, kernel_size=5, padding=0) # 54 -> 50
        self.act3  = torch.nn.ReLU()
        self.pool3 = torch.nn.AvgPool2d(kernel_size=2, stride=5) # 50 -> 10
        
        self.conv4 = torch.nn.Conv2d(
            in_channels=32, out_channels=64, kernel_size=5, padding=0) # 10 -> 6
        self.act4  = torch.nn.ReLU()
        self.pool4 = torch.nn.MaxPool2d(kernel_size=2, stride=2) # 6 -> 3
        
        self.fc1   = torch.nn.Linear(3 * 3 * 64, 120) # 3 * 3 * 128 -> 120
        self.act3  = torch.nn.ReLU()
        
        self.fc2   = torch.nn.Linear(120, 84)
        self.act4  = torch.nn.ReLU()
        
        self.fc3   = torch.nn.Linear(84, 10)
        self.act5  = torch.nn.ReLU()
                
        self.fc4   = torch.nn.Linear(10, 2)

    
    def forward(self, x):
        #   X -- ,   
        x = self.conv1(x)
        x = self.act1(x)
        x = self.pool1(x)
        
        x = self.conv2(x)
        x = self.act2(x)
        x = self.pool2(x)
        
        x = self.conv3(x)
        x = self.act3(x)
        x = self.pool3(x)
        
        x = self.conv4(x)
        x = self.act4(x)
        x = self.pool4(x)
        
        x = x.view(x.size(0), x.size(1) * x.size(2) * x.size(3))

        x = self.fc1(x)
        x = self.act3(x)
        x = self.fc2(x)
        x = self.act4(x)
        x = self.fc3(x)
        x = self.act5(x)
        x = self.fc4(x)
        return x
    
OlegNet5 = OlegNet5()
```
